```python
import pandas as pd
import matplotlib.pyplot as plt

# wczytanie danych z pliku Salary Data (2)
df = pd.read_csv('ML2025-1/Salary Data (2).csv', encoding='utf-8-sig')

# zgrupawanie wg wieku i obliczenie średniej dla pracowników w danym wieku
df_age_mean = df.groupby("Age")[["Salary"]].mean()

# stworzenie wykresu dla zgrupowanych danych
df_age_mean.plot(kind="bar", color="orange")

# dodanie etykiet, tytułu i siatki
plt.ylabel("Średnia pensja [$]")
plt.xlabel("Wiek [lat]")
plt.title("Średnia pensja wg. wieku pracowników")
plt.grid(axis="y")

# wyświetlenie wykresu
plt.show()
```
